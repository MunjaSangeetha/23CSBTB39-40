```python
import matplotlib.pyplot as plt

# Dataset
people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]
```

```python
# Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(age, height, color='blue', s=100, edgecolor='k')
plt.title('Scatter Plot: Height vs Age')
plt.xlabel('Age')
plt.ylabel('Height')
plt.grid(True)
plt.show()

# Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(people, weight, color='green', edgecolor='black')
plt.title('Bar Chart: Weight of People')
plt.xlabel('People')
plt.ylabel('Weight')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Histogram
plt.figure(figsize=(10, 6))
plt.hist(age, bins=4, color='purple', edgecolor='black')
plt.title('Histogram: Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
